## 1. __if 문__
가장 널리 사용되는 조건문으로 특정 조건에 맞으면 코드를 실행하고 그렇지 않으면 건너뛴다 <br>
아래의 예시가 가장 일반적인 형식:<br>
if <조건식1>:<br>
&nbsp;&nbsp;&nbsp;&nbsp;<문1> <br>
elif <조건식2>:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;<문2> <br>
else:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;<문3> <br>

위의 과정에서 <조건식1>이 참일경우 <문1>이 실행, <조건식2>가 참일 경우 <문2>가 실행, 그리고 <조건식1>과 <조건식2>가 모두 거짓일 경우에는 <문3>이 실행 된다
(들여쓰기에 주의하세요!)

#### 예시1 - if

```python
score = 90
if score >= 90:
    print("축하합니다")
```

#### 예시2 - if ~ else

```python
score = 80
if score >= 90:
    print("축하합니다")
else:
    print("노력하세요")
```

#### 예시3 - if ~ elif ~ else

```python
score = 70
if score >= 90:
    print("축하합니다")
elif score >= 80:
    print("노력하세요")
else:
    print("분발하세요")
```

## 2. __for 문__
가장 많이 사용하는 반복문이며 형식은 다음과 같다: <br>
for <타겟> in <객체>: <br>
&nbsp;&nbsp;&nbsp;&nbsp;<문1> <br>
<객체>의 각 항목이 <타겟>으로 치환되어 <문1>을 실행 <br>
<타겟>은 변수의 이름과 같이 작성자가 편의대로 지정 가능

```python
a = ['cat', 'cow', 'tiger']
for name in a:
    print(len(name), name)
```

for문의 반복을 멈추려면 'break'를 사용한다

```python
for i in range(10):
    print(i * 2)
    if i == 3:
        break
```
