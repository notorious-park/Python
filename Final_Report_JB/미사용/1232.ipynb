```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_excel("team_data/소비유형별.xlsx")
```

```python
df.head(5)
```

```python
df = df.T
tcard = df[0]
tcard
```

```python
type(tcard)
```

```python
tcard.index
```

```python
tcard = tcard[6:]
tcard.tail(5)
```

```python
type(tcard)
```

```python
tcard.name = '월별카드매출액'
```

```python
tcard.values
```

```python
frame = pd.DataFrame(tcard)
frame.tail(5)
```

```python
frame.index
```

```python
frame.index.names = ['기준년월']
frame.tail(5)
```

```python
frame = frame.reset_index()
frame
```

```python
frame['기준년월'].str.split('/').tolist()
```

```python
frame1 = pd.DataFrame(frame['기준년월'].str.split('/').tolist(),columns=['year', 'month'])
frame1.head(5)
```

```python
frame2 = pd.concat([frame1, frame], axis = 1)
frame2.head(5)
```

```python
frame2019 = frame2[frame2['year'] == '2019']
frame2020 = frame2[frame2['year'] == '2020']
frame2021 = frame2[frame2['year'] == '2021']

frame2019
```

```python
fig = plt.figure(figsize=(16, 10))
sns.set(style="white", context="talk", font="Malgun Gothic")

plt.plot(frame2019['month'], frame2019["월별카드매출액"], label="2019년도")
plt.plot(frame2020['month'], frame2020["월별카드매출액"], label="2020년도")
plt.plot(frame2021['month'], frame2021["월별카드매출액"], label="2021년도")
# plt.rc('font', family='Malgun Gothic')
plt.ylabel("월별개인신용카드승인금액(10조원)")
plt.legend()
plt.show()
```
