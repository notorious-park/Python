### 워드클라우드 시작

```python
import nltk
from konlpy.tag import Kkma
from konlpy.tag import Okt
```

```python
okt = Okt()
```

>국민의힘

```python
# 트위터 형태소 분석 후 명사만 추출
km_nouns = okt.nouns(km)
len(km_nouns)
```

```python
km_nouns[:10]
```

```python
one_words = []
tow_words = []
etc_words = []

for noun in km_nouns:
    if len(noun) == 1:
        one_words.append(noun)
    elif len(noun) == 2:
        tow_words.append(noun)
    else:
        etc_words.append(noun)
        
len(one_words), len(tow_words), len(etc_words)
```

```python
one_words = list(set(one_words))
tow_words = list(set(tow_words))
etc_words = list(set(etc_words))

len(one_words), len(tow_words), len(etc_words)
```

```python
for word in one_words:
    print(word, end='\t')
```

```python
stopwords = []
# stop_words = ['제', '월', '일','조','수','때','그','이','바','및','안']
stopwords.extend(one_words)
len(stopwords)
```

```python
# stop_words 빼기
km_nouns = [ word for word in km_nouns if word not in stop_words ]
len(km_nouns)
```

```python
km_nouns[:10]
```

```python
km  = nltk.Text(km_nouns, name='국민의힘')
type(km), len(km)
```

```python
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
# nltk패키지를 통해 konlpy에서 뽑은 명사들 상위 500개 추출
data = km.vocab().most_common(500)
len(data)
```

```python
# data를 딕셔너리 형태로 변환
dict_data = dict(data)
len(dict_data)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
from PIL import Image
from wordcloud import ImageColorGenerator
```

```python
# 이미지파일 numpy로 읽기
shape_img = np.array(Image.open('images/km.png'))
cloud_img = ImageColorGenerator(shape_img)
```

```python
wordcloud = WordCloud(font_path='C:/Windows/Fonts/malgun.ttf',
                     width = 300,
                     height = 300,
                     relative_scaling = 0.6,
                     mask = shape_img, 
                     background_color = 'white',
                     min_font_size=5, max_font_size=90).generate_from_frequencies(dict_data)

save_img = 'wordcloud/km_fin.png'
plt.figure(figsize=(15, 15))
plt.imshow(wordcloud.recolor(color_func=cloud_img))
plt.axis('off')
plt.savefig(save_img)
```
