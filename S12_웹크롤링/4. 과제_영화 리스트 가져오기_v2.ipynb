```python
from bs4 import BeautifulSoup 
from urllib.request import Request, urlopen
import re
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver'
```

```python
def get_read_web(url):    
    req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(req).read()
    soup = BeautifulSoup(html, "lxml")    
    return soup
```

```python
soup = get_read_web(url)
soup.title
```

```python
table_tags = soup.find_all('table')
len('table_tags')
```

```python
list_ranking = soup.find('table', 'list_ranking')
list_ranking
```

```python
tr_tags = list_ranking.find_all('tr')
len(tr_tags)
```

```python
tr_tags
```

```python
tr_tags[2]
```

```python
# 영화명
title = tr_tags[2].find("a").get_text()
title
```

```python
range_list = []

for tr_tag in range(1,11):
    range_ac1 = soup.find_all('td', 'ac')[tr_tag*3-2].find('img')['alt']
    print(range_ac1)
    if range_ac1 == 'down':
        comp = -1
    elif range_ac1 == 'up':
        comp = 1
    else:
        comp = 0
    range_list.append(comp)
        
range_list
```

```python
# url
link = tr_tags[2].find("a")['href']
link
```

```python
# 순위
rank = tr_tags[2].find('img')
# rank
rank = rank['alt']
rank = int(rank)
rank
```

```python
title_list = []
rank_list = []
url_list = []

for tr_tag in tr_tags[2:12]:
    rank = int(tr_tag.find('img')['alt'])
    rank_list.append(rank)
    
    title = tr_tag.find("a").get_text()
    title_list.append(title)
    
    url = tr_tag.find("a")['href']
    url_list.append('https://movie.naver.com/'+url)
    
```

```python
rank_list
```

```python
title_list
```

```python
url_list
```

```python
import pandas as pd

data = {'Rank' : rank_list, 'Movie' : title_list,  'Compare' : range_list, 'URL':url_list}
df = pd.DataFrame(data)
```
