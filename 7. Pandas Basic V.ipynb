### Summary statistics by level
> 단계별 요약통계

- level 옵션 : 어떤 한 축에 대해 합을 구하고 싶은 단계를 지정할 수 있는 옵션
- 내부적으로는 pandas의 groupby 기능을 이용해서 구현되었다.

```python
frame
```

```python
frame.sum(level = 'key2')   # 구
```

```python
frame.groupby(level=1).sum()   # 신 / level = 0 : key1     /     level = 1 : key2
```

```python
frame.sum(level = 'color', axis = 1)   # 구
```

```python
frame.groupby(level = 1, axis = 1).sum()   # 신 / level = 0 : state     /     level = 1 : color
```

### Using a DataFrame's columns
> DataFrame의 칼럼 사용하기

- DataFrame.set_index 함수 : 하나 이상의 칼럼을 색인으로 하는 새로운 DataFrame을 생성한다.
- DataFrame.reset_index 함수 : 계층적 샌인 단계가 칼럼으로 이동한다.

```python
frame = DataFrame({'a': range(7), 
                   'b': range(7, 0, -1),
                   'c': ['one', 'one', 'one', 'two', 'two', 'two', 'two'],
                   'd': [0, 1, 2, 0, 1, 2, 3]})
frame
```

```python
frame2 = frame.set_index(['c','d'])   # 변수 c,d의 값을 index로 지정  # drop default 값 True
frame2
```

```python
frame.set_index(['c','d'], drop = False)   # 변수 c,d를 남겨두고 index만 지정
```

```python
frame2.reset_index()
```
