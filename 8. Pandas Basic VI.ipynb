### Integer indexing
> 정수 색인

- pandas 객체를 정수로 색인해서 사용하는 일은 파이썬에서 리스트나 튜플 같은 기본 자료 구조에서 사용하는 색인의 의미와 약간 달라서 종종 실수를 할 수 있으므로 주의한다 !!

```python
ser = Series(np.arange(3.))
ser
```

```python
ser.iloc[-1]
```

```python
ser2 = Series(np.arange(3.), index = ['a','b','c'])
ser2
```

```python
ser2[-1]
```

```python
ser.loc[:1]      # loc : 해당숫자까지 추출
```

```python
ser.iloc[:1]    # iloc : 해당숫자 -1까지 추출
```

```python
ser3 = Series(range(3), index = [-5,1,3])
ser3
```

```python
ser3.iloc[2]
```

```python
frame = DataFrame(np.arange(6).reshape(3,2), index = [2,0,1])
frame
```

```python
frame.iloc[0]
```

#### Case2. Download Excel Loading

```python
# --------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------
# 엑셀 파일 읽어오기!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
df = pd.read_excel('./data/Stock_Corporations.xlsx')
df.head(10)
# --------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------
```

```python
df = df.sort_values(['상장일'], ascending = True)
df
```

```python
df.지역.unique()
# df['지역'].unique()   # 같은 결과
```

```python
df[df.지역 == '광주광역시'].head()
```

```python
mask = (df.지역 == '광주광역시') | (df.지역 == '전라남도') | (df.지역 == '전라북도')
honam_df = df[mask]
honam_df.head(10)
```

```python
honam_corp = honam_df[['회사명','종목코드']]
honam_corp
```

```python
honam_corp = honam_corp.rename(columns = {'회사명' : 'Corp.Honam', '종목코드' : 'Code.KS'})
honam_corp.head(10)
```

```python
# --------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------
# 엑셀 파일로 저장 방법!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
honam_corp.to_excel('./data/Honam_Stock_Corporations.xlsx')
# --------------------------------------------------------------------------------------------
# --------------------------------------------------------------------------------------------
```
