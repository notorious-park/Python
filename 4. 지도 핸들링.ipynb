### 서울시 구별 와아파이존 빈도수 시각화
>서울시 구별 무료와이파이존의 수를 지도에 나타낸다.

- 서울형 지도태깅 서울 Wi-Fi 정보
- 파일이름 : data/seoul_wifi_zone.csv
- Practice Data : [Download]

```python
import pandas as pd
import json
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
geo_path = 'data/skorea_municipalities_geo_simple.json'
geo_str  = json.load(open(geo_path, encoding='utf-8'))
```

```python
type(geo_str)
```

```python
# import pandas as pd
df = pd.read_csv('data/seoul_wifi_zone.csv')
df.head()
```

```python
df.columns = (['구', '유형', '위치명', '위도', '경도', '설치회사'])
df.head()
```

```python
freq_gu = df['구'].value_counts()
freq_gu
```

```python
freq_gu.describe()
```

```python
map = folium.Map(location=[37.5102, 126.982], zoom_start=11)
map
```

```python
map1 = folium.Map(location = [37.5102, 126.982], tiles = 'Stamen Toner', zoom_start = 11)

map1.choropleth(geo_data = geo_str,
                data = freq_gu,
                columns = ['gu', 'counts'],
                fill_color = 'YlGnBu',
                key_on = 'feature.id'
               )
map1
```

```python
map1 = folium.Map(location = [37.5102, 126.982], tiles = 'Stamen Toner', zoom_start = 11)

map1.choropleth(geo_data = geo_str,
                data = freq_gu,
                columns = ['gu', 'counts'],
                fill_color = 'PuRd',
                key_on = 'feature.id'
               )
map1
```

```python
map1 = folium.Map(location = [37.5102, 126.982], tiles = 'Stamen Toner', zoom_start = 11)

map1.choropleth(geo_data = geo_str,
                data = freq_gu,
                columns = ['gu', 'counts'],
                fill_color = 'BuPu',
                key_on = 'feature.id'
               )
map1
```
