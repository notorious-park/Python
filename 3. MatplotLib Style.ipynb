# 컬러와 스타일 사용자 정의
> 맷플롯립에서 컬러를 정의하는 방법

- 삼중(triplet) : 빨강, 파랑, 초록 성분은 [0, 1] 간격 안에 있어야 한다.
- 사중(quadruplet) : 삼중처럼 작동하며, 네번째 성분은 투명값을 정의한다.
- 사전 정의 이름 : b(파랑), g(초록), r(빨강), c(청록,Cyan), m(자몽,Magenta), y(노랑), k(검정), w(하양)

> Practice : 붓꽃(Iris) 데이터 분석
- 위사이트에서 피셔의 붓꽃 데이터 집합인 텍스트 파일로부터 배열을 불러온다.
- 꽃받침(sepal), 꽃잎(petal)의 길이와 너비에 따른 Iris 분류 : Setosa / Versicolour / Virginica

```python
%more ./data/iris.data
```

```python
import numpy as np
import matplotlib.pyplot as plt

data_file = './data/iris.data'
label_set = (
    b'Iris-setosa',
    b'Iris-versicolor',
    b'Iris-virginica',
)

def read_label(label):
    return label_set.index(label)

data = np.loadtxt(data_file,
                  delimiter = ',',
                  converters = {4: read_label})

color_set = ('.00', '.50', '.75')
color_list = [color_set[int(label)] for label in data[:,4]]
```

```python
data[:10]
```

```python
sorted(set(color_list))
```

```python
plt.scatter(data[:,0], data[:,1], color = color_list)
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
plt.scatter(data[:,2], data[:,3], color = color_list)
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```

```python
plt.scatter(data[:,0], data[:,2], color = color_list, edgecolors = '0.0')
plt.xlabel('petal length (cm)')
plt.ylabel('petal length (cm)')
plt.show()
```

```python
plt.scatter(data[:,0], data[:,2], color = color_list, edgecolors = '0.0')
plt.xlabel('petal width (cm)')
plt.ylabel('petal width (cm)')
plt.show()
```
