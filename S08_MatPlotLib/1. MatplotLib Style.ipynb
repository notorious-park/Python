### 2) 범례를 적용하기 위해 label을 적용하고 legend 실행

```python
plt.figure(figsize=(6,4))
plt.plot(t,  y, label='sin')
plt.plot(t, dy, label='cos')
plt.grid()
plt.legend()                # legend
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

### 3. 그래프 색상 바꾸기

```python
plt.figure(figsize=(6,4))
plt.plot(t,  y, label='sin')
plt.plot(t, dy, 'm', label='cos')    # color : ['b','g','r','c','m','y','k','w']
plt.grid()
plt.legend()                
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

### 4. 선 굵기 바꾸기

```python
plt.figure(figsize=(6,4))
plt.plot(t,  y, lw = 5, label='sin')       # linewidth
plt.plot(t, dy, 'm', label='cos')
plt.grid()
plt.legend()                
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

### 마커 적용

```python
t = np.arange(0,5,0.5)

plt.figure
plt.plot(t, t    , 'r-')   # r / b / g => 색상
plt.plot(t, t**2 , 'bs')    # 뒤에 - => 실선   /   -- => 점선
plt.plot(t, t**3 , 'g^')    # 뒤에 s => square(사각형)   /   ^ => 삼각형
plt.show()
```

```python
t = np.arange(0,5,0.5)

plt.figure
plt.plot(t, t    , 'r-', t, t**2, 'bs', t, t**3, 'g^')   
plt.show()
```

```python
# 기본 default는 3시 방향에서 data 왼쪽으로 이어짐
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes  = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # 특정 값 분리시키기

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)   # default는 3시 방향이므로 왼쪽으로 90도 회전(12시 방향)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### box 그래프 작성

```python
s1 = np.random.normal(loc = 0  , scale = 1.0, size = 1000)  # loc : mean / scale : sd / size : n
s2 = np.random.normal(loc = 5  , scale = 0.5, size = 1000)
s3 = np.random.normal(loc = 10,  scale = 2.0, size = 1000)
```

```python
plt.figure(figsize=(10,4))
plt.plot(s1, label = 's1')
plt.plot(s2, label = 's2')
plt.plot(s3, label = 's3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (10,4))
plt.boxplot((s1,s2,s3))
plt.grid()
plt.show()
```

### subplot 적용     # https://d3js.org/

```python
# ---------------------------------------------------------------------------------------------
# 암기는 아니나 이해하자!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
plt.figure(figsize = (16,8))    # 1. 도화지 크기를 결정한 후
plt.subplot(221)                # 2. 구역을 나눈다
plt.subplot(222)                # 221 : 2행 2열(4개)로 나누고 그 중 첫번째
plt.subplot(212)                # 212 : 2행 1열(2개)로 나누고 그 중 두번쩌
# Example)
# (16,8) 사이즈를 40이라고 치면
# 221 : 2행2열로 나누니 한개당 사이즈 10
# 212 : 2행1열로 나누니 한개당 사이즈 20
# ---------------------------------------------------------------------------------------------
```

```python
t = np.arange(0,5,0.01)

plt.figure(figsize = (16,8))
plt.subplot(411)
plt.plot(t,np.sqrt(t))

plt.subplot(423)
plt.plot(t,t**2)

plt.subplot(424)
plt.plot(t,t**3)

plt.subplot(413)
plt.plot(t,np.sin(t))

plt.subplot(414)
plt.plot(t,np.cos(t))
```
