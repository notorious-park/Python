# matplotlib을 이용한 데이터 시각화 프로그래밍

> 파이썬으로 구현하는 데이터 분석용 2D & 3D 그래프

```python
import matplotlib.pyplot as plt

X = range(100)
Y = [value ** 2 for value in X]

plt.plot(X, Y)
plt.show()
```

```python
# sin-1.py

import math
import matplotlib.pyplot as plt

T = range(100)
X = [(2 * math.pi * t) / len(T) for t in T]
Y = [math.sin(value) for value in X]

plt.plot(X, Y)
plt.show()
```

```python
# sin-2.py

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi, 100)
Y = np.sin(X)

plt.plot(X, Y)
plt.show()
```

```python
# 이항식 : x^2 - 2x + 1

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-3, 2, 200)
Y = X**2 - 2*X + 1.

plt.plot(X, Y)
plt.show()
```

> 다중 곡선 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

X  = np.linspace(0, 2*np.pi, 100)
Ya = np.sin(X)
Yb = np.cos(X)

plt.plot(X, Ya)
plt.plot(X, Yb)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_slope(X,Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys / Xs)

X = np.linspace(-3,3,100)
Y = np.exp(-X ** 2)

plt.plot(X,Y)
plot_slope(X,Y)
plt.show()
```

```python
import matplotlib.pyplot as plt

X, Y = [], []
for line in open('my_data.txt', 'r'):
    values = [float(s) for s in line.split()]
    X.append(values[0])
    Y.append(values[1])
    
plt.plot(X,Y)
plt.show()
```

```python
import matplotlib.pyplot as plt

with open('my_data.txt', 'r') as f:
    X, Y = zip(*[[float(s) for s in line.split()] for line in f])
    
plt.plot(X, Y)
plt.show()
```

> 분할 막대 차트 그리기

```python
import matplotlib.pyplot as plt

Y1 = [5., 30., 45., 22.]
Y2 = [5., 25., 50., 20.]

X = range(4)

plt.bar(X, Y1, color = 'blue')
plt.bar(X, Y2, color = 'red', bottom = Y1)    # 누적 막대차트 그리는 CASE!!!!!!!!!!!!!!!!!!!!!
plt.show()
```

```python
# 값을 두 집합 이상으로 분할
import numpy as np
import matplotlib.pyplot as plt

Y1 = np.array([5., 30., 45., 22.])
Y2 = np.array([5., 25., 50., 20.])
Y3 = np.array([10.,  20.,  10.,  10.])
X = range(4)

plt.bar(X, Y1, color='b')
plt.bar(X, Y2, color='g', bottom=Y1)
plt.bar(X, Y3, color='r', bottom=Y1+Y2)
plt.show()
```

> 점 그리기

```python
# [0,1] 간격에서 무작위로 그려진 좌표를 갖는 1024개의 점을 표시
import numpy as np
import matplotlib.pyplot as plt

data = np.random.rand(1024, 2)

plt.scatter(data[:,0], data[:,1])
plt.show()
```

> 상자그림 그리기

```python
# 정규분포에서 얻은 100개의 무작위 값에 대한 상자그림
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(100)

plt.boxplot(data)
plt.show()
```

> 삼각화 그리기

```python
# 공간 위치를 처리할 때 삼각화 생긴다. 점과 이웃 관계 사이의 거리를 보여줌
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri

data = np.random.rand(100, 2)

triangles = tri.Triangulation(data[:,0], data[:,1])

plt.triplot(triangles)
plt.show()
```
